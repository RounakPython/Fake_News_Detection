# Load Data

## Load Libraries

```python
# import the modules
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, confusion_matrix, f1_score, roc_auc_score
from wordcloud import WordCloud, STOPWORDS 
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import GradientBoostingClassifier,AdaBoostClassifier
from xgboost import XGBClassifier

import warnings
warnings.filterwarnings("ignore")

print('Setup Completed')
```

## Load Dataset

```python
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')

print(train.shape)
print(test.shape)
train.head()
```

# Exploratory Data Analysis

## Handling Null values

```python
# Checking null values
percentage_null_values = (train.isnull().sum()/train.isnull().count())*100

percentage_null_values
```

```python
train.isnull().sum()
```

```python
train.fillna("Unknown", inplace=True)
test.fillna("Unknown", inplace=True)
```

```python
train.isnull().sum()
```

```python
train.drop(columns=['id'], inplace=True)
test.drop(columns=['id'], inplace=True)
train.head()
```

## Data Processing

```python
# Code Starts here

# declare empty list 'corpus'
corpus=[]

# for loop to fill in corpus
for i in range(0,train.shape[0]):
    # retain alphabets
    review = re.sub('[^a-zA-Z]', ' ', train['text'].iloc[i] )
    # convert to lower case
    review=review.lower()
    # tokenize
    review=review.split()
    # initialize stemmer object
    ps=PorterStemmer()
    # perform stemming
    review=[ps.stem(word) for word in review if not word in set(stopwords.words('english'))]
    # join elements of list
    review=' '.join(review)
    # add to 'corpus'
    corpus.append(review)

# Code ends here
```

```python
fake_word_string = ' '.join([text for text in corpus])
```

## Ploting a Word cloud chart

```python
# Plot word cloud map

from wordcloud import WordCloud

wordcloud = WordCloud(width=800, height=500, random_state=21, max_font_size=110).generate(fake_word_string)
plt.figure(figsize=(10, 7))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis('off')
plt.show()
```

# Model Building

## Count Vectorization

```python
# Instantiate count vectorizer
cv = CountVectorizer(max_features=1500)

# Independent variable
X = cv.fit_transform(corpus).toarray()

# dependent variable
y = train['label']

print(X.shape, y.shape)
# Counts
count = y.value_counts()
print(count)
```

## Predefined functions

```python
models = {
          'Decision Tree Classifier (Gini)': DecisionTreeClassifier(criterion='gini'),
          'Decision Tree Classifier (Entropy)' : DecisionTreeClassifier(criterion='entropy'),
          'Random Forest Classifier':RandomForestClassifier(), 
          'Ada Boost Classifier' : AdaBoostClassifier(random_state=0),
          'XGBoost Classifier':XGBClassifier()
         }
```

```python
# Model Building

def model_building(model_values, X_data, y_data):
    '''
    calculate and returns Accuracy and ROC AUC Score for all models in a dataframe format
    '''
    X_train, X_test, y_train, y_test = train_test_split(X_data, y_data, random_state=0, test_size=0.2)
    
    classifier = []
    accuracy = []
    roc_auc_Score = []
    for name, model in model_values.items():
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)
        classifier.append(name)
        accuracy.append(accuracy_score(y_pred, y_test))
        roc_auc_Score.append(roc_auc_score(y_test, y_pred))
    
    results = pd.DataFrame({'Model':classifier, 'Accuracy Score':accuracy, 'ROC AUC Score':roc_auc_Score})
    print(results)        
```

## Model selection

```python
# Call Model building function

model_building(models, X, y)
```
